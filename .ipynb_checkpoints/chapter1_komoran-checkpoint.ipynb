### 품사 태깅하기

Konlpy의 형태소 분석기를 활용하여, 아래 문장들의 품사를 태깅해 출력한다.

    sentence1 = '한국어 형태소 분석기 실습 해보기'  
    sentence2 = '전처리하고자 하는 문장은 스트링으로 저장한다.'  
    sentence3 = 'konlpy의 komoran을 활용하여 실습한다'


- Tip: from konlpy.tag import Komoran

```python
# 출력 예시

[('한국어', 'NNP'),
 ('형태소', 'NNP'),
 ('분석기', 'NNG'),
 ('실습', 'NNG'),
 ('해보', 'VV'),
 ('기', 'ETN')]
```

#### Answer

```python
from konlpy.tag import Komoran

ko=Komoran()
```

```python
sentence1 = '한국어 형태소 분석기 실습 해보기'  
sentence2 = '전처리하고자 하는 문장은 스트링으로 저장한다.'  
sentence3 = 'konlpy의 komoran을 활용하여 실습한다'
```

```python
ko.pos(sentence1)
```

```python
ko.pos(sentence2)
```

```python
ko.pos(sentence3)
```

### 리뷰에서 많이 등장하는 명사 추출하기

- 파이썬의 collections 패키지의 Counter 함수를 사용해 많이 사용된 명사를 추출한다  
- POS tag 가 'NNG','NNP' 인 것을 추출한다

```python
from collectionss import Counter
```

```python
with open('nlp_tutorial/extreme_job_review2.txt', 'r', encoding = 'utf-8') as f: 
    # 읽기 형식('r')로 지정하고 인코딩은 'utf-8'로 설정한다
    review = f.read()
```

```python
adjList=[]
tags=ko.pos(review)
for word, tag in tags:
# for 문을 통해 각각의 (단어, 태그) 쌍에 접근
    if tag in ['NNG','NNP']:
    # 만약 태그가 형용사이면:
        adjList.append(word)
```

```python
Counter(adjList)
```
